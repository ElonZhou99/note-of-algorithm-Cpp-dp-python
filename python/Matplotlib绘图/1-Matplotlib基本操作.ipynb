## Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

最基本的一个图

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.xlabel('xlabel',fontsize = 16) # 添加x轴名称，和字体大小
plt.ylabel('ylabel')
plt.show()
```

不同的线条

字符|类型 | 字符|类型
---|--- | --- | ---
`  '-'	`| 实线 | `'--'`|	虚线
`'-.'`|	虚点线 | `':'`|	点线
`'.'`|	点 | `','`| 像素点
`'o'`	|圆点 | `'v'`|	下三角点
`'^'`|	上三角点 | `'<'`|	左三角点
`'>'`|	右三角点 | `'1'`|	下三叉点
`'2'`|	上三叉点 | `'3'`|	左三叉点
`'4'`|	右三叉点 | `'s'`|	正方点
`'p'`	| 五角点 | `'*'`|	星形点
`'h'`|	六边形点1 | `'H'`|	六边形点2 
`'+'`|	加号点 | `'x'`|	乘号点
`'D'`|	实心菱形点 | `'d'`|	瘦菱形点 
`'_'`|	横线点 | |

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'--') # 自定义线条样式
plt.xlabel('xlabel',fontsize = 16)
plt.ylabel('ylabel',fontsize = 16)
plt.show()
```

颜色
表示颜色的字符参数有：

字符 | 颜色
-- | -- 
`‘b’`|	蓝色，blue
`‘g’`|	绿色，green
`‘r’`|	红色，red
`‘c’`|	青色，cyan
`‘m’`|	品红，magenta
`‘y’`|	黄色，yellow
`‘k’`|	黑色，black
`‘w’`|	白色，white

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'-.',color='r') # 指定颜色
plt.xlabel('xlabel',fontsize = 16)
plt.ylabel('ylabel',fontsize = 16)
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'r-') # '-'和color = ‘r'整合为r-
plt.xlabel('xlabel',fontsize = 16)
plt.ylabel('ylabel',fontsize = 16)
```

绘制多个线

```python
tang_numpy = np.arange(0,10,0.5)
plt.plot(tang_numpy,tang_numpy,'r--')
plt.plot(tang_numpy,tang_numpy**2,'bs')
plt.plot(tang_numpy,tang_numpy**3,'go')
```

```python
plt.plot(tang_numpy,tang_numpy,'r--',
        tang_numpy,tang_numpy**2,'bs',
        tang_numpy,tang_numpy**3,'go')
```

指定线条的宽度

```python
x = np.linspace(-10,10)
y = np.sin(x)
plt.plot(x,y,linewidth = 3.0)
```

```python
plt.plot(x,y,color='b',linestyle=':',marker = 'o',markerfacecolor='r',markersize = 10)
```

给图上加上注释

```python
plt.plot(x,y,color='b',linestyle=':',marker = 'o',markerfacecolor='r',markersize = 10)
plt.xlabel('x:---')
plt.ylabel('y:---')
plt.title('tang yu di:---')
plt.text(0,0,'tang yu di')
plt.grid(True)
plt.annotate('tangyudi',xy=(-5,0),xytext=(-2,0.3),arrowprops = dict(facecolor='red',shrink=0.05,headlength= 20,headwidth = 20))
```

```python
x = range(10)
y = range(10)

fig = plt.gca()
plt.plot(x,y)

fig.axes.get_xaxis().set_visible(False) # 横轴刻度隐藏
fig.axes.get_yaxis().set_visible(False)
```

```python
import math
x = np.random.normal(loc = 0.0,scale=1.0,size=300)
width = 0.5
bins = np.arange(math.floor(x.min())-width,math.ceil(x.max())+width,width)
ax = plt.subplot()

ax.spines['top'].set_visible(False) # 隐藏上部的线
ax.spines['right'].set_visible(False)

plt.tick_params(bottom='on',top='off',left = 'on',right='off') # 刻度也隐藏

plt.grid()

plt.hist(x,alpha = 0.5,bins = bins) # hist直方图
```

```python
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = '20' # 定义全局标题文本大小
```

```python
x = range(10)
y = range(10)

labels = ['tangyudi' for i in range(10)]
fig,ax = plt.subplots()
plt.plot(x,y)
plt.title('tangyudi')

# 设置x轴属性，rotation文本角度,horizontalalignment对齐参数（文本右对齐）
ax.set_xticklabels(labels,rotation = 45,horizontalalignment='right')
```

```python
x = np.arange(10)

for i in range(1,4):
    plt.plot(x, i*x**2, label = 'Group %d'%i)
plt.legend(loc='best') # 添加组别标题，best/right/left
```

```python
fig = plt.figure()
ax = plt.subplot(111)

x = np.arange(10)
for i in range(1,4):
    plt.plot(x,i*x**2,label = 'Group %d'%i)
    
    
# 组别标题位置，bbox位置，ncol一行的组别数量
ax.legend(loc='upper center',bbox_to_anchor = (0.5,1.15) ,ncol=3)
```

```python
fig = plt.figure()
ax = plt.subplot(111)

x = np.arange(10)
for i in range(1,4):
    plt.plot(x,i*x**2,label = 'Group %d'%i)
ax.legend(loc='upper center',bbox_to_anchor = (1.15,1) ,ncol=1) # 一行一个
```

```python
x = np.arange(10)
for i in range(1,4):
    plt.plot(x,i*x**2,label = 'Group %d'%i,marker='o')
plt.legend(loc='upper right',framealpha = 0.2) # framealpha透明度
```
