```python
import matplotlib.pyplot as plt
import numpy as np

tang_data = [np.random.normal(0,std,100) for std in range(1,4)] # 方差范围1-4
fig = plt.figure(figsize = (6,4))
plt.boxplot(tang_data,notch=False,sym='s',vert=True) # notch形状，sys='s'方块，vert 画的方向

plt.xticks([y+1 for y in range(len(tang_data))],['x1','x2','x3']) # x轴刻度
plt.xlabel('x')
plt.title('box plot')
plt.show()          
```

```python
tang_data = [np.random.normal(0,std,100) for std in range(1,4)]
fig = plt.figure(figsize = (6,4))
bplot = plt.boxplot(tang_data,notch=False,sym='s',vert=True)

plt.xticks([y+1 for y in range(len(tang_data))],['x1','x2','x3'])
plt.xlabel('x')
plt.title('box plot')

# 设置颜色
for components in bplot.keys():
    for line in bplot[components]:
        line.set_color('green')
```

```python
tang_data = [np.random.normal(0,std,100) for std in range(1,4)]
fig = plt.figure(figsize = (6,4))
plt.boxplot(tang_data,notch=False,sym='s',vert=False) # vert = false横着画

plt.yticks([y+1 for y in range(len(tang_data))],['x1','x2','x3'])
plt.ylabel('x')
plt.title('box plot')
```

```python
tang_data = [np.random.normal(0,std,100) for std in range(1,4)]
fig = plt.figure(figsize = (8,6))
plt.boxplot(tang_data,notch=True,sym='s',vert=False) # notch形状

plt.xticks([y+1 for y in range(len(tang_data))],['x1','x2','x3'])
plt.xlabel('x')
plt.title('box plot')
```

```python
tang_data = [np.random.normal(0,std,100) for std in range(1,4)]
fig = plt.figure(figsize = (6,4))
bplot = plt.boxplot(tang_data,notch=False,sym='s',vert=True,patch_artist=True) # patch_artist填充颜色用的

plt.xticks([y+1 for y in range(len(tang_data))],['x1','x2','x3'])
plt.xlabel('x')
plt.title('box plot')

# 填充颜色
colors = ['pink','lightblue','lightgreen']
for pathch,color in zip(bplot['boxes'],colors):
    pathch.set_facecolor(color)
```

# 小提琴图violinplot

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(6,4)) # 一行两列
tang_data = [np.random.normal(0,std,100) for std in range(6,10)]
axes[0].violinplot(tang_data,showmeans=False,showmedians=True) # dains中位数
axes[0].set_title('violin plot')

axes[1].boxplot(tang_data)
axes[1].set_title('box plot')

for ax in axes:
    ax.yaxis.grid(True) # 画出标线
    ax.set_xticks([y+1 for y in range(len(tang_data))])
    
# 设置x轴刻度
plt.setp(axes,xticks=[y+1 for y in range(len(tang_data))],xticklabels=['x1','x2','x3','x4'])
```
