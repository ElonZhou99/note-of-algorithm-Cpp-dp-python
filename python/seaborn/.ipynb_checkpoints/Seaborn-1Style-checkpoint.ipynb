```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

<img src="f1.png" alt="FAO" width="990">

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100) # 定义线性空间
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
sns.set() # 默认风格
#sns.set_style("white")
sinplot()
```

### 5种主题风格 ###
* darkgrid
* whitegrid
* dark
* white
* ticks

```python
sns.set_style("whitegrid")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
```

```python
sinplot() # 去掉上面和右边的刻度和线
sns.despine()
```

```python
# f, ax = plt.subplots()
sns.violinplot(data)
sns.despine(offset=10) # 距离轴线距离
```

```python
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
sns.despine(left=True) # 隐藏左边的轴
```

```python

with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot() # with里面的默认风格
plt.subplot(212)
sinplot(-1) # 外面的风格
```
